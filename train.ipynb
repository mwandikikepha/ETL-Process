```python
# importing requests and pandas
import requests
import pandas as pd
```

```python
# reading the csv file
df = pd.read_csv("uber.csv")
df.head()
```

```python
# checking how many nulls are there
df.isnull().sum()
```

```python
# Finding all the columns with numeric data types(int, float)
numeric_columns = df.select_dtypes(include = ['number']).columns
numeric_columns
```

```python
# Replacing nulls in numeric_columns with 0
df[numeric_columns] = df[numeric_columns].fillna(0)
```

```python
# Checking all the numeric_columns
df[numeric_columns]
```

```python
# Finding all the columns with object data types
char_columns = df.select_dtypes(include = ['object']).columns
char_columns
```

```python
# Checking all the char_columns
df[char_columns]
```

```python
# Replacing all the zero in char_columns with 'unknown'
df[char_columns] = df[char_columns].fillna('unknown')
df[char_columns]
```

```python
# Choosing the columns you want from all this data
wanted_columns = ['Date', 'Booking ID', 'Customer ID', 'Pickup Location', 'Drop Location', 'Ride Distance', 'Ride_Category']
```

```python
ride_col = df['Ride Distance']
ride_col
```

```python
#Adding a new column "Ride_Category" which classifies the ride as long, medium or short
def Ride_Category(ride_col):
    if 1 < ride_col < 5:
        return 'low'
    elif 6 < ride_col < 20:
        return 'medium'
    elif 21 < ride_col < 50:
        return 'high'
    else:
        return 'No distance'
```

```python
df['Ride_Category'] = df['Ride Distance'].apply(Ride_Category)
```

```python
# Creating a new dataframe 'WantedCols'
WantedCols = df[wanted_columns]
WantedCols.head()
```

```python
from sqlalchemy import create_engine
```

```python
# creating a connection engine to postgresql db hosted on aiven
cleaned_data = create_engine ('')
```

```python
#Pushing the cleaned data to postgresql. Pushes it as a table named CleanedUberData, to a schema named public
WantedCols.to_sql("CleanedUberData", cleaned_data, index= False, if_exists='replace', schema = 'public')
```

```python


# read back from PostgreSQL to confirm if 'CleanedUberData' has been pushed there
check = pd.read_sql("SELECT * FROM public.\"CleanedUberData\" LIMIT 5;", cleaned_data)
print(check)
```
